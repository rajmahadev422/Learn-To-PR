## Derivatives

```python
c = ((1 + smp.sin(x))/(1 - smp.cos(x)))**2
c
```

```python
smp.diff(c, x)
```

```python
d = (smp.log(x)/smp.log(5))**(x/2)
d
```

```python
f = smp.diff(d, x)
f
```

```python
f.subs(x, 2)
```

```python
g, h = smp.symbols('g h', cls=smp.Function)
g = g(x)
h = h(x + g)
h
```

```python
smp.diff(h, x)
```

## Basic Antideratives

```python
m = smp.csc(x)*cot(x)
m
```

```python
smp.integrate(m, x)
```

```python
m = 2/smp.root(1 - x**2, 2) - 1/x**smp.Rational(1/4)
m
```

```python
smp.integrate(m, x)
```

### Finding c

```python
eq = 8*x + smp.csc(x)**2
eq
```

```python
m = smp.integrate(eq, x)
m
```

```python
c = -7 - m.subs(x, smp.pi/2)
y = m + c
y
```

```python
y.subs(x, smp.pi/2) # check initial condition
```

## Complex Equations

```python
f = (1 + smp.sqrt(x))**(smp.Rational(1,3))/smp.sqrt(x)
f
```

```python
smp.integrate(f, x)
```

## Definite Integral

```python
f = smp.exp(x)/smp.sqrt(smp.exp(2*x) + 9)
f
```

```python
smp.integrate(f, (x, 0, smp.log(4)))
```

```python
f = (x**10)*smp.exp(x)
f
```

```python
t = smp.symbols('t')
```

```python
y = smp.integrate(f, (x, 1, t))
y
```

```python
y.subs(t, 0)
```

## Improper integrals

```python
f = 16*smp.atan(x)/(1 + x**2)
f
```

```python
smp.integrate(f, (x, 0, smp.oo))
```

## Sequence and Series

```python
n = smp.symbols('n')
```

```python
f = 6/(4**n)
f
```

```python
smp.Sum(f, (n, 0, smp.oo)).doit()
```

```python
f = 2**(n+1)/5**n
f
```

```python
smp.Sum(f, (n, 0, smp.oo)).doit()
```

```python
f = smp.atan(n)/n**(1.1)
f
```

```python
smp.Sum(f, (n, 1, smp.oo)).doit().n()
```
